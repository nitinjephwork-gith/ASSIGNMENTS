```python
import pandas as pd
```

```python
date_strings = ['2026-01-01', '2026-01-02', '2026-01-03']

timeseries = pd.to_datetime(date_strings)

print(timeseries)

ts = pd.Series([100, 200, 300], index=timeseries)

print(ts)
```

```python
df1 = pd.DataFrame({
    'ID': [1, 2, 3, 4],
    'Name': ['Aman', 'Riya', 'Karan', 'Neha']
})

df2 = pd.DataFrame({
    'ID': [2, 3, 4, 5],
    'City': ['Delhi', 'Mumbai', 'Pune', 'Jaipur']
})

print(df1)
print(df2)
```

```python
inner_merge = pd.merge(df1, df2, on='ID', how='inner')

print(inner_merge)
```

```python
left_join = pd.merge(df1, df2, on='ID', how='left')

print(left_join)
```

```python
right_join = pd.merge(df1, df2, on='ID', how='right')

print(right_join)
```

```python
df1_index = df1.set_index('ID')
df2_index = df2.set_index('ID')

index_join = df1_index.join(df2_index)

print(index_join)
```

```python
df3 = pd.DataFrame({
    'ID': [1, 1, 2, 2],
    'Dept': ['HR', 'IT', 'HR', 'IT'],
    'Salary': [50000, 60000, 55000, 65000]
})

df4 = pd.DataFrame({
    'ID': [1, 2, 2],
    'Dept': ['HR', 'HR', 'IT'],
    'Bonus': [5000, 4000, 6000]
})

multi_merge = pd.merge(df3, df4, on=['ID', 'Dept'])

print(multi_merge)
```

```python
df_a = pd.DataFrame({
    'ID': [1, 2],
    'Name': ['Aman', 'Riya']
})

df_b = pd.DataFrame({
    'ID': [3, 4],
    'Name': ['Karan', 'Neha']
})

df_c = pd.DataFrame({
    'ID': [1, 2, 3, 4],
    'Marks': [85, 90, 88, 92]
})

print(df_a)
print(df_b)
print(df_c)
```

```python
concat_df = pd.concat([df_a, df_b], ignore_index=True)

print(concat_df)
```

```python
final_merge = pd.merge(concat_df, df_c, on='ID')

print(final_merge)
```
