```python
data = {
    'Name': ['Aman', 'Riya', 'Karan'],
    'Age': [21, 22, 20]
}

df = pd.DataFrame(data)

for index, row in df.iterrows():
    print(index, row['Name'], row['Age'])



for row in df.itertuples():
    print(row.Name, row.Age)


for col in df:
    print(col)
```

```python
result = df[df['Age'] > 20]

print(result)
```

```python
result = df[(df['Age'] > 20) & (df['Name'] == 'Riya')]

print(result)
```

```python
data = {
    'Name': ['Aman', 'Riya'],
    'Age': [21, 22]
}

df = pd.DataFrame(data)

new_row = pd.DataFrame({
    'Name': ['Karan'],
    'Age': [20]
})

df = pd.concat([df.iloc[:1], new_row, df.iloc[1:]]).reset_index(drop=True)

print(df)
```
