```python

def maximum(a, b, c):
    return max(a, b, c)


x = int(input("Enter first number: "))
y = int(input("Enter second number: "))
z = int(input("Enter third number: "))

print("Maximum number is:", maximum(x, y, z))
```

```python


def distinct_list(lst):
    return list(set(lst))

numbers = [1, 2, 2, 3, 4, 4, 5]

print("Original List:", numbers)
print("Distinct List:", distinct_list(numbers))
```

```python
def multiply_list(lst):
    result = 1

    for num in lst:
        result = result * num

    return result

numbers = [2, 3, 4, 5]

print("Multiplication Result:", multiply_list(numbers))
```

```python
def factorial(n):
    fact = 1

    for i in range(1, n + 1):
        fact = fact * i

    return fact

num = int(input("Enter a number: "))

print("Factorial =", factorial(num))
```

```python
def reverse_string(text):
    return text[::-1]

string = input("Enter a string: ")

print("Reversed String:", reverse_string(string))
```

```python

def check_range(num, start, end):
    if num >= start and num <= end:
        print(num, "is within the range")
    else:
        print(num, "is outside the range")

number = int(input("Enter number: "))

check_range(number, 1, 100)
```

```python

def even_numbers(lst):

    print("Even Numbers are:")

    for num in lst:
        if num % 2 == 0:
            print(num)

numbers = [1, 2, 3, 4, 5, 6, 7, 8]

even_numbers(numbers)
```

```python
def is_prime(n):

    if n <= 1:
        return False

    for i in range(2, n):
        if n % i == 0:
            return False

    return True

num = int(input("Enter a number: "))

if is_prime(num):
    print(num, "is a Prime Number")
else:
    print(num, "is not a Prime Number")
```

```python
def count_case(text):

    upper = 0
    lower = 0

    for ch in text:

        if ch.isupper():
            upper += 1

        elif ch.islower():
            lower += 1

    print("Uppercase letters:", upper)
    print("Lowercase letters:", lower)

string = input("Enter a string: ")

count_case(string)
```
