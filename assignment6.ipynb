```python
import random


uscore = 0
cscore = 0

choices = ["rock", "paper", "scissor"]


for i in range(5):

    print("\nRound", i + 1)

    
    uch = input("Enter rock/paper/scissor: ").lower()

    
    cch = random.choice(choices)

    print("Computer chose:", cch)

    
    if uch == cch:
        uscore += 1
        cscore += 1
        print("Match Draw")

    elif ((uch == "rock" and cch == "scissor") or
          (uch == "paper" and cch == "rock") or
          (uch == "scissor" and cch == "paper")):

        uscore += 2
        print("You Win")

    else:
        cscore += 2
        print("Computer Wins")


print("\nFinal Scores")
print("Your Score =", uscore)
print("Computer Score =", cscore)

if uscore > cscore:
    print("Congratulations! You won the game.")
elif cscore > uscore:
    print("Computer won the game.")
else:
    print("Game Draw")
```
