```python
import csv

file_name = "address_book.csv"
```

```python
fields = ["Name", "Address", "Mobile", "Email"]

records = []


n = int(input("How many records do you want to enter? "))
```

```python
for i in range(n):
    print(f"\nEnter details for Record {i+1}")

    name = input("Enter Name: ")
    address = input("Enter Address: ")
    mobile = input("Enter Mobile Number: ")
    email = input("Enter Email: ")

    records.append([name, address, mobile, email])
```

```python
with open(file_name, "w", newline="") as file:
    writer = csv.writer(file)

   
    writer.writerow(fields)

    
    writer.writerows(records)

print("\nCSV file created successfully!")
```

```python
print("\nStored Address Book Records:\n")

with open(file_name, "r") as file:
    reader = csv.reader(file)

    for row in reader:
        print(row)
```

```python
import sqlite3


conn = sqlite3.connect("college.db")

cur = conn.cursor()
```

```python

cur.execute("""
CREATE TABLE Student (
    id INTEGER PRIMARY KEY,
    name TEXT,
    age INTEGER,
    course TEXT
)
""")

cur.execute("""
CREATE TABLE Teacher (
    id INTEGER PRIMARY KEY,
    name TEXT,
    subject TEXT
)
""")

print("Tables created successfully!")
```

```python
cur.execute("INSERT INTO Student VALUES (1, 'Nitin', 20, 'CSE')")
cur.execute("INSERT INTO Student VALUES (2, 'Rahul', 21, 'IT')")
cur.execute("INSERT INTO Student VALUES (3, 'Aman', 19, 'ECE')")

cur.execute("INSERT INTO Teacher VALUES (1, 'Sharma Sir', 'Python')")
cur.execute("INSERT INTO Teacher VALUES (2, 'Mehta Sir', 'DBMS')")

conn.commit()

print("Records inserted successfully!")
```

```python
print("\nAll Student Records:")
cur.execute("SELECT * FROM Student")

for row in cur.fetchall():
    print(row)

print("\nStudents with age > 19:")
cur.execute("SELECT * FROM Student WHERE age > 19")

for row in cur.fetchall():
    print(row)
```

```python

cur.execute("""
UPDATE Student
SET course = 'AI'
WHERE id = 2
""")

conn.commit()

print("\nRecord updated successfully!")

print("\nUpdated Student Table:")
cur.execute("SELECT * FROM Student")

for row in cur.fetchall():
    print(row)
```

```python

cur.execute("""
DELETE FROM Student
WHERE id = 3
""")

conn.commit()

print("\nRecord deleted successfully!")

print("\nFinal Student Table:")
cur.execute("SELECT * FROM Student")

for row in cur.fetchall():
    print(row)

conn.close()
```
