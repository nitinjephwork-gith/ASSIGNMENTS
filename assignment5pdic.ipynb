```python
# Q1 Repeat Tuple

t = (1, 2, 3)
print("Q1:", t * 3)
```

```python
# Q2 Join Tuples

t1 = (1, 2)
t2 = (3, 4)
t3 = (5, 6)
print("Q2:", t1 + t2 + t3)
```

```python
# Q3 Check Element

t = (10, 20, 30, 40)
print("Q3:", 30 in t)

```

```python
# Q4 Total, Highest, Lowest

t = (12, 45, 7, 89, 23)

total = 0
highest = t[0]
lowest = t[0]

for num in t:
    total += num

    if num > highest:
        highest = num

    if num < lowest:
        lowest = num

print("Q4 Total =", total)
print("Q4 Highest =", highest)
print("Q4 Lowest =", lowest)

```

```python
# Q5 Filter Values > 10

n = (3, 14, 7, 22, 9, 41, 18, 5)

result = ()

for i in n:
    if i > 10:
        result += (i,)

print("Q5:", result)

```

```python
# Q6 Count Set Elements

s = {"cat", "dog", "bird", "fish"}

count = 0
for item in s:
    count += 1

print("Q6:", count)

```

```python
# Q7 Union of Sets

s1 = {1, 2, 3}
s2 = {3, 4, 5}

print("Q7:", s1 | s2)

```

```python
# Q8 Intersection

s1 = {1, 2, 3, 4}
s2 = {3, 4, 5, 6}

print("Q8:", s1 & s2)

```

```python
# Q9 Symmetric Difference

print("Q9:", s1 ^ s2)
```
