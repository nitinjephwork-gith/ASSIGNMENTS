```python
import numpy as np

arr = np.array([
    [6, -8, 73, -110],
    [np.nan, -8, 0, 94]
])

print("Original Array:")
print(arr)


arr = np.nan_to_num(arr, nan=0)

print("\nNaN Replaced with 0:")
print(arr)


transpose = arr.T

print("\nTranspose:")
print(transpose)
```

```python

arr = np.arange(24).reshape(2, 3, 4)

print("Original Shape:")
print(arr.shape)

new_arr = np.moveaxis(arr, 0, 2)

print("New Shape:")
print(new_arr.shape)

print(new_arr)
```

```python

arr = np.array([
    [10, np.nan, 30],
    [20, 25, np.nan],
    [30, 35, 50]
])

print("Original Array:")
print(arr)

col_mean = np.nanmean(arr, axis=0)

indices = np.where(np.isnan(arr))

arr[indices] = np.take(col_mean, indices[1])

print("\nAfter Replacing NaN:")
print(arr)
```

```python

arr = np.array([
    [10, -5, 20],
    [-8, 30, -12],
    [15, -4, 25]
])

print("Original Array:")
print(arr)

arr[arr < 0] = 0

print("\nAfter Replacing Negative Values:")
print(arr)
```
