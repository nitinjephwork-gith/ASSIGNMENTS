```python
import numpy as np

arr = np.array([10, 20, 30, 40, 50])

print("Array:", arr)
print("Shape:", arr.shape)
print("Dimension:", arr.ndim)
print("Data Type:", arr.dtype)
print("Item Size:", arr.itemsize)
```

```python

arr = np.array([1, 2, 3, 4, 5, 6])

arr2d = arr.reshape(2, 3)

print(arr2d)
```

```python

arr = np.full((3, 3), 9)
print(arr)
```

```python

arr = np.linspace(25, 125, 10)
print(arr)
```

```python

lst = [5, 10, 15, 20]
arr = np.array(lst)
print(arr)
```

```python

arr = np.array([10, 20, 30, 40, 50])

print("Original:", arr)
print("Reversed:", arr[::-1])
```

```python

arr = np.arange(48).reshape(4, 4, 3)

print(arr)

print("\nRequired Value:")
print(arr[1, 0, 2])
```

```python

arr = np.arange(1, 17).reshape(4, 4)

print("Original Array:")
print(arr)

result = arr[::2, 1::2]

print("\nOdd Rows and Even Columns:")
print(result)
```

```python

arr = np.arange(48).reshape(4, 4, 3)

result = arr[1, :2, :2]
print(result)
```

```python

arr = np.array([
    [23, 56, 78, 93],
    [71, 82, 13, 24]
])

for i in range(arr.shape[0]):
    for j in range(arr.shape[1]):
        if arr[i][j] % 2 != 0:
            arr[i][j] = -1

print(arr)
```

```python

arr = np.array([1, 0, 2, 0, 3, 0, 4])

indices = np.nonzero(arr)
print(indices)
```

```python

arr1 = np.array([10, 20, 30])
arr2 = np.array([5, 15, 25])

print("Addition:")
print(arr1 + arr2)

print("Multiplication:")
print(arr1 * arr2)
```

```python

arr1 = np.array([15, 20, 25])
arr2 = np.array([10, 40, 37])

dot_product = np.dot(arr1, arr2)

print("Dot Product =", dot_product)
```
